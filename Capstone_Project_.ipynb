```python
import pandas as pd

# Membaca file CSV
df = pd.read_csv('retail_warehouse_sales.csv')

# Tampilkan 5 baris pertama
df.head()
```

```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby(['YEAR', 'MONTH'])[['RETAIL SALES', 'WAREHOUSE SALES']].sum().reset_index()
monthly_sales['PERIOD'] = monthly_sales['YEAR'].astype(str) + '-' + monthly_sales['MONTH'].astype(str).str.zfill(2)

plt.figure(figsize=(12,6))
plt.plot(monthly_sales['PERIOD'], monthly_sales['RETAIL SALES'], label='Retail Sales')
plt.plot(monthly_sales['PERIOD'], monthly_sales['WAREHOUSE SALES'], label='Warehouse Sales')
plt.xticks(rotation=45)
plt.title('Monthly Retail & Warehouse Sales')
plt.xlabel('Period')
plt.ylabel('Sales')
plt.legend()
plt.tight_layout()
plt.show()
```
